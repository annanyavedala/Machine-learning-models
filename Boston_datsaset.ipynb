```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
from sklearn.datasets import load_boston
boston= load_boston()
```

```
boston
```

```
print(boston['DESCR'])
```

```
print(boston['data'])
```

```
print(boston['target'])
```

```
features= pd.DataFrame(boston.data, columns= boston.feature_names)
```

```
labels=pd.DataFrame(boston.target, columns=['MEDV'])
```

```
labels
```

```
features.isnull().sum()
```

```
data= pd.concat([features, labels], axis=1)
```

```
FEATURES= list(features.columns)
LABEL='MEDV'
```

```
import seaborn as sns
def plot_scatter(df, features, label):
  plt.figure(figsize=(20,20))
  for loc, feature in enumerate(features):
    ax= plt.subplot(4,4,loc+1)
    sns.scatterplot(x=feature, y=label, data=df)
  return None
```

```
plot_scatter(data, FEATURES, LABEL)
```

```
sns.heatmap(data.corr())
```

```
def plot_hist(df, features):
  plt.figure(figsize=(20,20))
  for loc, feature in enumerate(features):
    ax= plt.subplot(4,4,loc+1)
    sns.distplot(df[feature],bins=100, color='#581645')
  return None
```

```

plot_hist(data, data.columns)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(features, labels, test_size=0.33, random_state=0)
X_train.shape
```

```
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train.values, y_train.values)
```

```
model.coef_
```

```
model.intercept_
```

```
plt.scatter(data['CRIM'], data['MEDV'])
plt.plot(data['CRIM'], model.coef_[0][0]*data['CRIM']+ model.intercept_,)
```

```
test_pred=model.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
import math
math.sqrt(mean_squared_error( y_test, test_pred))
```
